# An introduction to the Convolutional Neural Network (CNN)

A Convolutional Neural Network is a model which is designed to process data which is structured with grid-like topology. Essentually, a CNN provides a bespoke architecture specifically created to handle image processing [1]. Unlike traditional neural networks which are more general-purpose, CNN's are designed to exploit the grid structure of images to construct the features and characteristic of that image as we propogate through the network [2][3]. CNN implimentation has proved succesfull in many application domains such as facial recognition [3], medical diagnosis [4] and voice recognition [5].

## 1. The Convolution layers

These layers are responsbile for perfoming the operation of convolution. This operation functions as follows:

- suppose we have an image of a given length (L), width (W) and height (H, number of colour channels e.g. RGB images would have a height of 3)
- a small subsection of the image is considered and a filter is applied to it (essentially a small neural network). The output of this neural network is propogated into the next layer where the hidden neurons in this next layer essentially only recieve information about this particular small region of the image. Please see [3] for a formal convolutional formula for one pixel.
- this filter is slid across the entire input image and the process is repeated in what are known as strides [1].

This operation is important as each filter can work to produce a feature map which inspects different aspects of an image such as the presence of edges and shapes [2]. Additionally, the implimentation means that instead of connecting every pixel of the image to each neuron which can lead to a very high number of connections, the filter can learn to detect a specific pattern in a small area of the image and this information is passed only to a small subset of neurons in the hidden layer meaning fewer connections working to reduce the number of model parameters and making the model increasingly efficient. 


## 2. Pooling layers

The pooling layers are responsbible for ensuring translation invariance by reducing the spartial size of the feature maps (the outputs from our filters) through down-sampling. This means that even if objects are in different positions with the photo, they can still be recognizes and extracted as features by the model. In terms of image classification, pooling is akin to reducing the resolution (width*height in pixels) [3]. The above diagram mentions a specific type of pooling known as max pooling. Max pooling works by taking a feature map (for example, a 4 pixels x 4 pixels map) and choosing a window size (for example, 2 pixels x 2 pixels). In each window, the pixel with the largest output from the feature map is kept and the rest are removed. See the below diagram for a visual description of this process. Furtermore, alternative techniques such as average pooling and mixed pooling exist (where both max pooling and average pooling take place).

## 3. Dense layers

These fully connected at the end of the neural network are responsbile for processing the features extracted by the convolutional layers in previous layers. As aforementioned, the convolutional layers can extract edges and shapes that make up larger images (e.g. a face). The dense layers are responsible for combining these low level features into a high-level representation which allows the network to fulfill its purpose of classification. Since these layers are fully connected, they can learn the complex relationships between the low-level features extracted by the Convolutional layers. Since the output of multiple iterations of concolutional layers and pooling layers is again a feature map, which we can think of as a matrix, the dense layers flattern the map into a 1-dimensional vector before being passed to the output layer.  


## 4. Output layers

In an image classification task scenario, this layer would include the same number of neurons as there are classes in the dataset. For example, if there were 4 class categories then there would be 4 neurons in the output layer.
